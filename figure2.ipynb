```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# PBMC-full

```python
pbmc_full = pd.read_excel('cell_count.xlsx', sheet_name='PBMC-full', index_col=0)
pbmc_full.head()
```

```python
sns.set(style="white")

g = pbmc_full.plot.bar(stacked=True, width=0.8)    #绘制堆积柱状图
sns.despine()
plt.xlabel('')
plt.ylabel('Number of cells', fontsize=16)
plt.xticks(fontsize=14, rotation=90)
plt.yticks(fontsize=14)
plt.legend(fontsize=14)
plt.title('Cell count of the PBMC-full dataset', fontsize=16)
plt.savefig('figures/Figure2A_PBMC_full.pdf', format='pdf', dpi=300, bbox_inches='tight')
```
